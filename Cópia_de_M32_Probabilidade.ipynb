```python
# Probabilidade - Tarefa | M32


import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Exercício 1

# Máquina embala pacotes com distribuição Normal: média=15kg, dp=0,1kg
# P(X > 15.2) = ?

prob_maior_152 = 1 - stats.norm.cdf(15.2, loc=15, scale=0.1)

print("=" * 55)
print("Exercício 1")
print("=" * 55)
print(f"P(X > 15,2 kg) = {prob_maior_152:.6f}")
print(f"Em porcentagem: {prob_maior_152 * 100:.4f}%")


# Exercício 2

# Percentil de ordem 95% para o peso dos pacotes

percentil_95 = stats.norm.ppf(0.95, loc=15, scale=0.1)

print("\n" + "=" * 55)
print("Exercício 2")
print("=" * 55)
print(f"Percentil 95%: {percentil_95:.4f} kg")
print(f"Ou seja, 95% dos pacotes pesam menos que {percentil_95:.4f} kg")


# Exercício 3

# Y1 ~ Normal Padrão (média=0, variância=1)
# Y2 ~ t-Student com 5 graus de liberdade
# Calcular a CDF (P(Yi < y)) para cada valor de Y

Y = np.linspace(-6, 6, 200)

cdf_normal = stats.norm.cdf(Y, loc=0, scale=1)  # CDF da Normal Padrão
cdf_t5     = stats.t.cdf(Y, df=5)               # CDF da t-Student (df=5)

print("\n" + "=" * 55)
print("Exercício 3")
print("=" * 55)
print("Primeiros 5 valores de Y:", Y[:5].round(3))
print("CDF Normal Padrão (primeiros 5):", cdf_normal[:5].round(6))
print("CDF t-Student df=5 (primeiros 5):", cdf_t5[:5].round(6))


# Exercício 4

# Visualização gráfica das CDFs do exercício 3

plt.figure(figsize=(10, 5))
plt.plot(Y, cdf_normal, label='Normal Padrão  N(0,1)', color='steelblue', linewidth=2)
plt.plot(Y, cdf_t5,     label='t-Student  df=5',       color='tomato',    linewidth=2, linestyle='--')

plt.title('Função de Distribuição Acumulada (CDF)', fontsize=14, fontweight='bold')
plt.xlabel('y', fontsize=12)
plt.ylabel('P(Y ≤ y)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0.5, color='gray', linestyle=':', linewidth=1)
plt.axvline(0,   color='gray', linestyle=':', linewidth=1)
plt.tight_layout()
plt.show()

print("Gráfico gerado!")


# Exercício 5
# Inversa da CDF (PPF / percentil) para Y1 e Y2
# P = valores de probabilidade de 0 a 1

P = np.linspace(0, 1, 20)

ppf_normal = stats.norm.ppf(P, loc=0, scale=1)  # inversa da Normal
ppf_t5     = stats.t.ppf(P, df=5)               # inversa da t-Student

print("\n" + "=" * 55)
print("Exercício 5")
print("=" * 55)
print("Primeiros 5 valores de P:", P[:5].round(4))
print("Percentil Normal (primeiros 5):", ppf_normal[:5].round(4))
print("Percentil t-Student df=5 (primeiros 5):", ppf_t5[:5].round(4))


# Exercício 6

# Gráfico dos percentis do exercício 5

# Remove -inf e +inf (extremos de probabilidade 0 e 1)
mask = np.isfinite(ppf_normal) & np.isfinite(ppf_t5)

plt.figure(figsize=(10, 5))
plt.plot(P[mask], ppf_normal[mask], label='Normal Padrão  N(0,1)', color='steelblue', linewidth=2)
plt.plot(P[mask], ppf_t5[mask],     label='t-Student  df=5',       color='tomato',    linewidth=2, linestyle='--')

plt.title('Inversa da CDF — Função Quantil (PPF)', fontsize=14, fontweight='bold')
plt.xlabel('Probabilidade p', fontsize=12)
plt.ylabel('Quantil Q(p)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.axhline(0, color='gray', linestyle=':', linewidth=1)
plt.axvline(0.5, color='gray', linestyle=':', linewidth=1)
plt.tight_layout()
plt.show()

print("Gráfico gerado!")
print("\n Todos os exercícios concluídos!")
```

**Conclusão Resumida**

Neste notebook, foram exploradas as distribuições Normal e t-Student por meio do cálculo de probabilidades, percentis e da análise gráfica de suas funções CDF e PPF.

Os resultados mostraram que a probabilidade de um pacote pesar mais de 15,2 kg é de aproximadamente 2,28%, enquanto 95% dos pacotes possuem peso inferior a 15,1645 kg.

A comparação entre as distribuições evidenciou que a t-Student possui caudas mais pesadas do que a Normal, atribuindo maior probabilidade a eventos extremos. Essa característica também foi observada nos gráficos de percentis, nos quais a t-Student apresenta valores mais distantes da média para as mesmas probabilidades extremas.

De modo geral, a distribuição Normal é mais adequada para grandes amostras e dados estáveis, enquanto a t-Student é recomendada para amostras menores e cenários com maior incerteza. Além disso, quanto maiores os graus de liberdade da t-Student, mais ela se aproxima da distribuição Normal.
